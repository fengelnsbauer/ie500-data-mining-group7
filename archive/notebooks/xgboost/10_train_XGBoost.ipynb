```python
# train_xgboost.py

import logging
import sys
sys.path.append('../../')
import numpy as np
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from common.data_preparation import load_and_preprocess_data, prepare_features_for_xgboost, split_data_by_race
from models.xgboost.xgboost_model import F1DataPreprocessor, save_model_with_preprocessor
# Ensure you can import from parent directories


def train_xgboost_model():
    # Load and preprocess data
    df = load_and_preprocess_data()
    
    # Split data
    train_df, test_df = split_data_by_race(df, test_size=0.2, random_state=42)
    
    # Prepare features
    X_train, y_train = prepare_features_for_xgboost(train_df)
    X_test, y_test = prepare_features_for_xgboost(test_df)
    
    # Initialize preprocessor
    preprocessor = F1DataPreprocessor()
    
    # Fit and transform data
    X_train_scaled = preprocessor.fit_scalers(X_train, y_train)
    X_test_scaled = preprocessor.transform_data(X_test)
    
    # Set XGBoost parameters
    params = {
        'objective': 'reg:squarederror',
        'eval_metric': 'rmse',
        'max_depth': 6,
        'learning_rate': 0.1,
        'n_estimators': 1000,
        'subsample': 0.8,
        'colsample_bytree': 0.8
    }
    
    # Create DMatrix
    dtrain = xgb.DMatrix(X_train_scaled, label=y_train)
    dtest = xgb.DMatrix(X_test_scaled, label=y_test)
    
    # Train model
    model = xgb.train(
        params,
        dtrain,
        num_boost_round=1000,
        evals=[(dtrain, 'train'), (dtest, 'test')],
        early_stopping_rounds=50,
        verbose_eval=100
    )
    
    # Save model and preprocessor
    save_model_with_preprocessor(model, preprocessor, 'models/xgboost/xgboost_model.pkl')
    
    return model, preprocessor

train_xgboost_model()
```
