```python
import pandas as pd
import matplotlib.pyplot as plt

# Data preparation
data = {
    "layers": [2, 4, 5, 6, 6, 5, 3, 5, 2],
    "neurons_per_layer": [54, 75, 119, 33, 106, 125, 63, 109, 47],
    "test_loss": [
        4333.8969, 5251.6081, 4965.5721, 5426.6431, 5952.6115, 4988.8382,
        4781.6806, 5134.4355, 4790.5850
    ]
}

df = pd.DataFrame(data)

# Plotting test_loss vs neurons_per_layer
plt.figure(figsize=(15, 5), facecolor="black")

# Plot 1: Test loss vs Neurons per layer
plt.subplot(1, 3, 1, facecolor="black")
plt.scatter(df["neurons_per_layer"], df["test_loss"], color="blue", label="Loss")
plt.title("Test Loss vs Neurons Per Layer", color="white")
plt.xlabel("Neurons Per Layer", color="white")
plt.ylabel("Test Loss", color="white")
plt.grid(True, color="gray")
plt.legend(facecolor="black", edgecolor="white", labelcolor="white")

# Set the color of tick labels to white
plt.xticks(color="white")
plt.yticks(color="white")

# Plot 2: Test loss vs Layers
plt.subplot(1, 3, 2, facecolor="black")
plt.scatter(df["layers"], df["test_loss"], color="green", label="Loss")
plt.title("Test Loss vs Layers", color="white")
plt.xlabel("Layers", color="white")
plt.ylabel("Test Loss", color="white")
plt.grid(True, color="gray")
plt.legend(facecolor="black", edgecolor="white", labelcolor="white")

# Set the color of tick labels to white
plt.xticks(color="white")
plt.yticks(color="white")

# Plot 3: Test loss vs Layers * Neurons per Layer
df["layers_times_neurons"] = df["layers"] * df["neurons_per_layer"]
plt.subplot(1, 3, 3, facecolor="black")
plt.scatter(df["layers_times_neurons"], df["test_loss"], color="red", label="Loss")
plt.title("Test Loss vs Layers * Neurons Per Layer", color="white")
plt.xlabel("Layers * Neurons Per Layer", color="white")
plt.ylabel("Test Loss", color="white")
plt.grid(True, color="gray")
plt.legend(facecolor="black", edgecolor="white", labelcolor="white")

# Set the color of tick labels to white
plt.xticks(color="white")
plt.yticks(color="white")

# Adjust layout
plt.tight_layout()
plt.show()
```
