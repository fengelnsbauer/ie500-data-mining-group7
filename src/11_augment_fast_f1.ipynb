```python
import pandas as pd
import fastf1

fastf1.Cache.enable_cache("../data/cache")

df = pd.read_csv('../data/processed/export_v1.csv')

races_df = pd.read_csv('../data/raw_data/races.csv')

df.head()
```

```python
df = df[['raceId']]
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

df = pd.merge(df, races_df, on='raceId', how='left')

print(len(df))
df.head()
```

```python
races = []
for idx, row in df.iterrows():
    try:
        fp1 = fastf1.get_session(year=row['year'], gp=row['name'], identifier='fp1')
    except:
        fp1 = None
    try:
        fp2 = fastf1.get_session(year=row['year'], gp=row['name'], identifier='fp2')
    except:
        fp2 = None
    try:
        fp3 = fastf1.get_session(year=row['year'], gp=row['name'], identifier='fp3')
    except:
        fp3 = None

    races.append({
        'raceId': row['raceId'],
        'fp1_session': fp1,
        'fp2_session': fp2,
        'fp3_session': fp3
    })
```

```python

race = races[1100]['fp1_session']
race.load(laps=True)
```

```python
laps = race.laps
quick_laps = laps.pick_quicklaps()

quick_laps.head()
```

```python
quick_laps.loc[:, 'LapTime'] = quick_laps['LapTime'].dt.total_seconds().astype(float) * 1000
quick_laps.loc[:, 'Sector1Time'] = quick_laps['Sector1Time'].dt.total_seconds().astype(float) * 1000
quick_laps.loc[:, 'Sector2Time'] = quick_laps['Sector2Time'].dt.total_seconds().astype(float) * 1000




quick_laps.head()
```

```python
grp = quick_laps.groupby("Driver").mean(numeric_only=True)

grp
```
