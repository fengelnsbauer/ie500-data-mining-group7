```python
import pandas as pd

full_file_path = '../data/processed/export_combined_v1.csv'
df = pd.read_csv(full_file_path)

df.head()
```

```python
from sklearn.model_selection import train_test_split

# Define the target variable
y = df["positionOrder"]
X = df.drop(columns=["positionOrder"])

# Perform stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, shuffle=True, random_state=42
)

print(len(X_train[X_train["year"] >= 2018]), len(X_test[X_test["year"] >= 2018]))

# Display the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```
